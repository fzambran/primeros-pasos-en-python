### ¿Qué es una función?

<img src="https://cdn.ttgtmedia.com/rms/onlineImages/software_quality-dry_vs_wet_code.png" alt="" width="500">

Una función es un bloque de código con un nombre asociado, que recibe cero o más argumentos como entrada, sigue una secuencia de sentencias, la cuales ejecuta una operación deseada y devuelve un valor y/o realiza una tarea, este bloque puede ser llamados cuando se necesite.

<a id="funciones-built-in"></a>
### Funciones `built-in`
Se refiere a que ya vienen incorporadas en el Core de Python. Algunas son:
* `abs(v)`      : entrega valor absoluto de un valor.
* `all(x)`      : retorna `True` si todos los valores de un iterable son `True`.
* `any(x)`      : retorna `True` si algún valor de un iterable es `True`.
* `bool(v)`     : retorna True al menos que el valor ingresado sea vacío, `False`, `None` o 0.
* `float(v)`    : transforma valor a número de punto flotante (a decimal).
* `int(v)`      : transforma valor a número entero.
* `list(x)`     : crea un objeto tipo lista.
* `pow(b,e)`    : calcula $b^e$ (alternativa a `b**e`).
* `print(o)`    : convierte a `string` y muestra en pantalla un objeto.
* `range(i,f,s)`: crea una secuencia de números enteros desde `i` hasta `f-1` con salto `s`.
* `round(v, n)` : redondea un valor a `n` decimales.
* `str(o)`      : convierte un objecto a `string`.
* `type(o)`        : retorna tipo de un objeto.

```python
# Ejemplos de funciones built-in:
lista_booleanos = [True, False, False, False, True]
print(all(lista_booleanos))
print(any(lista_booleanos))
print(list(range(1, 11)))
print(list(range(10, 201, 10)))
```

```python
print(list("hola"))
print(list(range(10, 201, 10)))
```

### Funciones definidas por el usuario
<img src="https://naps.com.mx/blog/wp-content/uploads/2020/06/Funciones-en-Python.-Estructura-de-una-funci%C3%B3n.png" alt="" width="500">


* Para resolver problemas específicos, cuya solución no se encuestra implementada, necesitaremos crear nuestras propias funciones.
* Una función es una bloque de código que es ejecutado solo cuando se llama.
* Podemos crear funciones a través de la sentencia `def`.
* Retomemos el ejemplo de procesamiento del discurso:

```python
# Ejemplo: Crear funcion llamada 'limpieza texto'  que reciba una cadena de texto
# y que retorne una lista con las palabras 'limpias'.
def limpieza_texto(texto):
  discurso_lista = texto.split()
  palabras_limpias = []

  for palabra in discurso_lista:
    palabra_limpia = palabra.lower()
    palabra_limpia = palabra_limpia.strip("!,.¿?") # Eliminar caracteres del final o principio
    palabras_limpias.append(palabra_limpia)

  return palabras_limpias 
```

```python
discurso_1 = "Gracias por su entusiasmo, por su dedicación al trabajo, por su apertura al cambio, por sus aportes y por secundarme en tantas ideas, propuestas y proyectos, que hoy vemos realizados y que serán los pilares sobre los que se siga apoyando el desarrollo futuro de nuestro programa."
discurso_2 = "Nos pidieron que  dijéramos algunas palabras hoy, esta mañana en que de a poquito, empiezan a despedirse de Madrigal. Han sido muchos años, meses días y horas que han pasado todos ustedes juntos aquí, y no nos dimos ni cuenta, cuando de tanto ir y venir todos los días, que llegaba este día."
discurso_3 = "¿Han logrado los museo cumplir con su rol social?  ¿Cómo se comprometen nuestras instituciones frente a los cambios de la sociedad? El museo, ¿se ha convertido, realmente, en un espacio provocativo?  ¿Cuáles han sido los resultados? Sus acciones ¿han sido instrumentos de cambio"
```

```python
print("discurso 1:", limpieza_texto(discurso_1))
print("discurso 2:", limpieza_texto(discurso_2))
print("discurso 3:", limpieza_texto(discurso_3))
```

```python
print("-".join(  limpieza_texto(discurso_3)  ))
```

<a id="funciones-anonimas"></a>
### Funciones anónimas
* A diferencia de las funciones creadas con la sentencia `def`, las funciones anónimas no reciben un nombre.
* Se pueden crear a través de la palabra reservada `lambda`.
* Son útiles para crear funciones que resuelvan problemas sencillos.

```python
def calcular_imc(kg, m):
  imc = kg / m**2
  return imc
```

```python
# Ejemplo: Crear función que calcule el IMC (IMC = kg / m^2).
calcular_imc = lambda kg, m: kg / m**2
```

```python
print(calcular_imc(70, 2.2))
```

<a id="librerias-y-frameworks"></a>
### Librerías y frameworks
<img src="https://i.ibb.co/c8FPC9j/imgonline-com-ua-Big-Pic-25a1y-GPOw5ourl9g.png" width="500">


* Generalmente las funcionalidades que trae la base de python no es suficiente para nuestras necesidades, o desarrollar una función que nos ayude, eventualmente tomaría demasiado tiempo.
* Para ello podemos acceder a librerías o frameworks, que nos proporcionaran distintas funcionalidades a medida que las vayamos necesitando.
* __Framework__: Cuando necesitamos desarrollar un proyecto, con una estructura definida, y que la compatibilidad de sus funcionalidades esté asegurada, es recomendable usar un Framework, que es una colección de funcionalidades que siguen una misma norma o filosofía. Un ejemplo de framework es `django`, usado para desarrollo web.
* __Librería__: Es un conjunto más reducido de funcionalidades, con el objetivo de resolver un problema en concreto. A diferencia de los frameworks, el uso de librerías nos permite mayor libertad para elegir herramientas que resuelvan determinado problema. Por ejemplo, para ciencias de datos, comúnmente se utilizan en conjunto `numpy`, `pandas`, `matplotlib`, `scikit-learn`, `scipy`, `statsmodels`, etc.
* No todo es 100% framework o 100% librería.

<a id="manipulacion-de-arrays"></a>
### Manipulación de Arrays
* Los `array`s presentan muchos métodos que hacen que su manipulación sea más facil.
* Algunos métodos disponibles: `array.mean`, `array.sum`, `array.min`, `array.max`, `array.median`, `array.std`, `array.transpose`, `array.sort`, etc.
* Además, podemos realizar operaciones elemento a elemento entre `array`s con la misma dimensión, como la suma, resta, multiplicación y división.
* Consultar en la documentación todas las funciones y métodos disponibles.

```python
array_1 = np.array([10, 20, 40, 50, 60, 70, 80, 90, 100])
array_2 = np.array([5, 8, 1, 3, 9, 7, 2, 4, 0])

# Calcular la suma de todos los elementos de 'array_1'
print(array_1.sum())

# Calcular la media de todos los elementos de 'array_2'
print(array_2.mean())

# Multiplique 'array_1' y 'array_2'.
# Luego obtenga la desviación estándar del array resultante
print(array_1.shape)
print(array_2.shape)
multiplicacion = array_1 * array_2
print(multiplicacion)
print(multiplicacion.std()) # Desviación estándar
```

* Mediana: Percentil 50 de los datos, en numpy usamos función `median` de NumPy.

```python
print(np.median(array_1))
```

<a id="vectorizacion-de-arrays"></a>
### Vectorización de Arrays

* Es cuando una operación o función aplicada a un `array` es reflejada en todos sus valores internos.
* Esto supone una gran ventaja, ya que no tenemos que iterar para aplicar una función a cada valor de los `array`s. 

```python
lista_valores = [20, 50, 23, 60, 2]
print([valor + 2 for valor in lista_valores])
```

```python
# Ejemplo 1: Divida entre 10 el array_1, luego restar 1
print(array_1)
print(array_1 / 10 - 1)

# Ejemplo 2: Cree una función que convierta de grados farenheit a celsius
# C° = (F° - 32) / 1.8
# Luego convierta a grados celsius la siguiente lista de grados farenheit
farenheit = np.array([32, 40, 50, 60, 70, 80])
fa_a_ce = lambda f: (f - 32) / 1.8
print("farenheit:", fa_a_ce(farenheit))
```

```python
####### Importar abajo 'load_digits'
from sklearn.datasets import load_digits
#######
# Descomentar:
X = load_digits()["data"].reshape(1797, 8, 8)
```

* El dataset `X`, contiene 1797 imágenes en blanco y negro con dimensiones 8x8, cada imagen representa un número del 0 al 9. Muestre la primera imágen y la número 1300. ¿A qué número pertenece cada una?

* Un preprocesamiento básico de imágenes para entrenar un algoritmo de `Machine Learning` es *escalar* las imágenes al rango 0-1. Realice esto dividiendo el set de datos por su valor máximo y verifique viendo el mínimo y el máximo del nuevo `array`.

```python
# Responda aqui
X_proc1 = X / X.max()
print("rango:", X_proc1.min(), X_proc1.max())
```

* Otro tipo de preprocesamiento es *normalizar* el set de datos, esto es restarle la media y luego dividir por la desviación estándar el set de datos original. Cree una función que normalice los datos y luego apliquela al set de datos `X`. Finalmente verifique la media y desviación estándar de los datos resultantes. 

```python
# Responda aquí
normalizar = lambda x: (x - x.mean()) / x.std()
X_proc2 = normalizar(X)
print("media:", X_proc2.mean().round(5))
print("desviación estándar:", X_proc2.std().round(5))
```
